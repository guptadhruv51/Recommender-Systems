```python
import pandas as pd
import numpy as np
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
movies_df=pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies_df.head()
```

```python
credits.info()
```

```python
movies_df.info()
```

```python
credits_renamed=credits.rename(index=str,columns={'movie_id':'id'})
```

```python
movies_df_merge=movies_df.merge(credits_renamed,on='id')
movies_df_merge.head()
```

```python
movies_cleaned=movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_companies'],axis=0)
movies_cleaned.head()
```

```python
movies_cleaned.info()
```

```python
movies_cleaned['vote_average'].mean()
```

```python
v=movies_cleaned['vote_count']
R=movies_cleaned['vote_average']
C=movies_cleaned['vote_average'].mean()
m=movies_cleaned['vote_count'].quantile(0.70)
```

```python
movies_cleaned['Weighted_Average']=((R*v)+(C*m))/(v+m)
```

```python
movies_cleaned.head()
```

```python
movies_sorted=movies_cleaned.sort_values('Weighted_Average',ascending=False)
movies_sorted[['original_title','popularity','vote_count','Weighted_Average','vote_average']].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
sns.barplot(movies_sorted['Weighted_Average'][:10],y=movies_sorted['original_title'][:10])
plt.xlim(4,8.5)
plt.ylabel('Best Movies')
plt.xlabel('Weighted Average')
```

```python
popularity=movies_cleaned.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(popularity['popularity'][:10],popularity['original_title'][:10])
plt.ylabel('Movies')
plt.xlabel('Popularity')
```

Recommendation Based on Scaled Weighted Average And Popularity(50% to Both)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
weighted_df=scale.fit_transform(movies_cleaned[['Weighted_Average','popularity']])
```

```python
weighted_normalize=pd.DataFrame(weighted_df,columns=['Weighted_Average','Populairty'])
weighted_normalize.head()
```

```python
movies_cleaned[['Normalized_Weight_Avg','Normalized_Pop']]=weighted_normalize
```

```python
movies_cleaned.head()
```

```python
movies_cleaned['Score']=movies_cleaned['Normalized_Weight_Avg']*0.5 +movies_cleaned['Normalized_Pop']*0.5
movies_cleaned=movies_cleaned.sort_values('Score',ascending=False)
movies_score=movies_cleaned[['original_title','Normalized_Weight_Avg','Normalized_Pop','Score']]
movies_score.head()
```
