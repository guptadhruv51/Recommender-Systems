```python
Rating_count=pd.DataFrame(ratings.groupby('ISBN')['Ratings'].count())
Rating_count.sort_values(by='Ratings',ascending=False).head()
```

## To ensure Statistical significance, users with less than 200 ratings and books with less than 100 ratings are excluded

```python
counts1=ratings['User_ID'].value_counts()
ratings=ratings[ratings['User_ID'].isin(counts1[counts1>=200].index)]
count=ratings['Ratings'].value_counts()
ratings=ratings[ratings['Ratings'].isin(count[count>=100].index)]
```

```python
ratings_pivot=ratings.pivot_table(index='User_ID',columns='ISBN',values='Ratings')
userID = ratings_pivot.index
ISBN = ratings_pivot.columns
print(ratings_pivot.shape)
ratings_pivot.head()
```
